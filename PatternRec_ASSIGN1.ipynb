```python
from google.colab import files
import pandas as pd
import numpy as np
from scipy.spatial import distance
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
from io import StringIO
import math

# Upload files to Colab
uploaded_files = files.upload()


file_name = next(iter(uploaded_files))
df = pd.read_csv(StringIO(uploaded_files[file_name].decode('utf-8')))



numeric_predictors = df.select_dtypes(include='number').columns
numeric_predictors = numeric_predictors.drop('quality')


euclidean_results = {}
dotproduct_results={}
cosine_similarity_results = {}
pearson_correlation_results = {}

for predictor in numeric_predictors:
    euclidean_dist_per_row = math.dist(df[predictor] , df['quality'])
    euclidean_results[predictor] = np.mean(euclidean_dist_per_row)

    dotproduct_per_row= np.dot(df[predictor] , df['quality'])
    dotproduct_results[predictor]=(dotproduct_per_row)

    cosine_sim = 1 - cosine(df[predictor], df['quality'])
    cosine_similarity_results[predictor] = cosine_sim

    pearson_corr = df[predictor].corr(df['quality'])
    pearson_correlation_results[predictor] = pearson_corr


print("Euclidean Distance Ranking:")
for key in sorted(euclidean_results, key=euclidean_results.get):
    print(f"{key}: {euclidean_results[key]}")
print("=============")
print("Cosine Similarity Ranking:")
for key in sorted(cosine_similarity_results, key=cosine_similarity_results.get, reverse=True):
    print(f"{key}: {cosine_similarity_results[key]}")
print("=============")
print("Dot Product Ranking:")
for key in sorted(dotproduct_results, key=dotproduct_results.get, reverse=True):
    print(f"{key}: {dotproduct_results[key]}")
print("=============")
print("Pearson Correlation Ranking:")
for key in sorted( pearson_correlation_results, key= pearson_correlation_results.get, reverse=True):
    print(f"{key}: { pearson_correlation_results[key]}")
```

```python
from scipy.spatial.distance import euclidean, cosine
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import LabelEncoder

numeric_predictors = df.select_dtypes(include='number').columns
numeric_predictors = numeric_predictors.drop('quality')

# Encode 'type' column to numerical values
label_encoder = LabelEncoder()
df['type_encoded'] = label_encoder.fit_transform(df['type'])

type_euclidean_results = {}
type_cosine_results = {}
type_dot_product_results = {}
type_pearson_results = {}

for predictor in numeric_predictors:
    print(predictor)
    euclidean_dist = euclidean(df[predictor], df['type_encoded'])
    type_euclidean_results[predictor] = euclidean_dist

    cosine_sim = cosine_similarity(df[predictor].values.reshape(1, -1), df['type_encoded'].values.reshape(1, -1))[0, 0]
    type_cosine_results[predictor] = cosine_sim

    dot_product = np.dot(df[predictor], df['type_encoded'])
    type_dot_product_results[predictor] = dot_product

    pearson_corr = np.corrcoef(df[predictor], df['type_encoded'])[0, 1]
    type_pearson_results[predictor] = pearson_corr


print("Euclidean Distance Ranking:")
for key in sorted(type_euclidean_results, key=type_euclidean_results.get):
    print(f"{key}: {type_euclidean_results[key]}")
print("=============")
print("Cosine Similarity Ranking:")
for key in sorted(type_cosine_results, key=type_cosine_results.get, reverse=True):
    print(f"{key}: {type_cosine_results[key]}")
print("=============")
print("Dot Product Ranking:")
for key in sorted(type_dot_product_results, key=type_dot_product_results.get, reverse=True):
    print(f"{key}: {type_dot_product_results[key]}")
print("=============")
print("Pearson Correlation Ranking:")
for key in sorted(type_pearson_results, key=type_pearson_results.get, reverse=True):
    print(f"{key}: {type_pearson_results[key]}")
```
